سیستم پیشنهاد دهنده، به طور کلی سیستمی است که بر اساس سوابق کاربر، پیشنهادهایی ارائه کند که ممکن است کاربر را جذب کند

یکی از روش های پیدا کردن شباهت، پیدا کردن زاویه بین دو بردار هست. از هر چیزی میتوانیم بردار بسازیم. برای مثال بردار غذا خوردن بین ساعت های مختلف روز. یا بردار میانگین حرف زدن یک فرد در یک ساعت

برای پیدا کردن زاویه بین دو بردار میتوانیم، کسینوس بین دو بردار را حساب کنیم. هر چه این مقدار به یک نزدیک تر باشن یعنی زاویه بین دو بردار به صفر نزدیک تر است و در نتیجه شباهت بیشتری دارند
